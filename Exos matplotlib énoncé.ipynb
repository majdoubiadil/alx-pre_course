## Exercice 6 : Ecrivez un programme Python pour tracer plusieurs lignes avec différents styles de format en une seule commande en utilisant des tableaux.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 1, 4, 9, 16, 25])
y2 = np.array([0, 1, 2, 3, 4, 5])
plt.plot(x, y1, 'r-o', x, y2, 'b-^')
plt.xlabel("Axe X")
plt.ylabel("Axe Y")
plt.title("Plusieurs courbes avec styles différents")
plt.legend(["y = x²", "y = x"])
```

## Exercice 7 : Ecrire une programmation Python pour afficher un histogramme de la popularité des langages de programmation.
### Exemple de données:
* Langages de programmation: Java, Python, PHP, JavaScript, C #, C ++
* Popularité: 22.2, 17.6, 8.8, 8, 7.7, 6.

```python
import numpy as np
import matplotlib.pyplot as plt
x=['Java', 'Python', 'PHP', 'JavaScript', 'C #', 'C ++']
popularity=[22.2, 17.6, 8.8, 8, 7.7, 6]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color='blue')
plt.xlabel("languages")
plt.ylabel("popularity")
plt.title("Popularity of programing Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x)
```

## Exercice 8 : Ecrire un programme Python pour créer un diagramme à barres des scores par groupe et sexe. Utilisez plusieurs valeurs X sur le même graphique pour les hommes et les femmes.
### Exemple de données:
* Moyennes (hommes) = (22, 30, 35, 35, 26)
* Moyennes (femmes) = (25, 32, 30, 35, 29)

```python
import matplotlib.pyplot as plt
import numpy as np
groupes = ['G1', 'G2', 'G3', 'G4', 'G5']
moy_hommes = [22, 30, 35, 35, 26]
moy_femmes = [25, 32, 30, 35, 29]
x = np.arange(len(groupes))
largeur = 0.35
plt.bar(x - largeur/2, moy_hommes, width=largeur, label='Hommes')
plt.bar(x + largeur/2, moy_femmes, width=largeur, label='Femmes')
```

## Exercice 9 : Ecrivez un programme Python pour créer un histogramme horizontal avec des couleurs ordonnées différemment.
### Exemple de jeu de données:
* languages = [['Language', 'Science', 'Math'], ['Science', 'Math', 'Language'], ['Math', 'Language', 'Science']]
* valeurs = [{'Language': 75, 'Science': 88, 'Math': 96}, {'Language': 71, 'Science': 95, 'Math': 92}, {'Language': 75, 'Science': 90, 'Math': 89}]

```python
import numpy as np
import matplotlib.pyplot as plt
languages = [['Language', 'Science', 'Math'],
             ['Science', 'Math', 'Language'],
            ['Math', 'Language', 'Science']]
valeurs = [{'Language': 75, 'Science': 88, 'Math': 96},
           {'Language': 71, 'Science': 95, 'Math': 92},
           {'Language': 75, 'Science': 90, 'Math': 89}]
plt.xlabel("languages")
plt.ylabel("valeurs")
colors = {'Language': 'blue','Science': 'red','Math': 'green'}
for i in range(3):
    plt.barh(i, valeurs[i][languages[i][0]], color=colors[languages[i][0]])
    plt.barh(i + 0.2, valeurs[i][languages[i][1]], color=colors[languages[i][1]])
    plt.barh(i + 0.4, valeurs[i][languages[i][2]], color=colors[languages[i][2]])





```
