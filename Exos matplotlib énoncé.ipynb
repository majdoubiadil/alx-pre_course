## Exercice 1 : Ecrivez un programme Python pour dessiner une ligne avec une étiquette appropriée sur l'axe x, l'axe y et un titre.

```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5, 6]
y=[3, 6, 9, 12, 15, 18]
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Diagramme de mesure")
```
